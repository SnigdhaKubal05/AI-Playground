```python
from transformers import pipeline

task = input("Enter task number (1-4): ")

if task == "1":
    classifier = pipeline("sentiment-analysis")
    print(classifier("I love programming"))

elif task == "2":
    sentence = pipeline("sentiment-analysis")
    print(sentence("I love Snigdha"))

elif task == "3":
    generator = pipeline("text-generation", model="gpt2")
    print(generator("I want to become a developer", max_length=30))

elif task == "4":
    question = pipeline("question-answering")
    context = "Python is a programming language widely used in AI nowadays."
    print(question(question="What is Python?", context=context))
```
